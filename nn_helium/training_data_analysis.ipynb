```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
regular_top = [74, 75, 73, 72, 72]
regular_avg = [71.5, 71.2, 72, 69.5, 70]

heavy_top = [66, 66, 66, 65, 65]
heavy_avg = [64.2, 65, 64, 63.8, 64]
x = np.arange(1, 6)
```

```python
plt.plot(x, regular_top, label="Red top")
plt.plot(x, regular_avg, label="Red avg")
plt.plot(x, heavy_top, label="Heavy top")
plt.plot(x, heavy_avg, label="Heavy avg")
plt.grid()
plt.legend()
plt.title("Bowling speeds June-July")
plt.xlabel("Session number")
plt.ylabel("mph")
plt.xticks(x)
```
